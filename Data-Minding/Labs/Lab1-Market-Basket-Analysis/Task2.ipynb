```python
#import libraries
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = pd.read_excel('Egyptian_Grocery_Transactions.xlsx',header=None)
dataset
```

```python
dataset.shape
```

```python
dataset.columns = ['Items']
dataset
```

```python
transactions = []
items = dataset['Items'].values
for i in range (0, len(items)):
    transactions.append(items[i].split(","))

transactions           
```

```python
encoder = TransactionEncoder()
transactionsEncoded = encoder.fit(transactions).transform(transactions)
transactionsEncoded = pd.DataFrame(data = transactionsEncoded, columns=encoder.columns_, dtype=int)
transactionsEncoded
```

```python
support = apriori(transactionsEncoded, min_support=0.1, use_colnames=True)
support.sort_values(by = 'support')
```

```python
AnalysisData = association_rules(support, metric='confidence', min_threshold=0.3)
AnalysisData[['antecedents', 'consequents', 'support', 'confidence', 'lift']].sort_values(by = 'lift')

#There is no weak rules and all rules with lift greater than 1 is strong rule
```
